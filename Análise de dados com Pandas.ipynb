## Selecionando as cinco cidades com mais novos casos em 31/07/2021


```python
df = df[['city', 'newCases']]
df5 = df.nlargest(5, 'newCases')
df5
```

```python
df5.plot('city', 'newCases', title='5 cidades com mais novos casos em 31/7/2021')
plt.xlabel('Cidades')
plt.ylabel('Número de novos casos')
```

## Selecionando as dez cidades com mais novos casos em 31/07/2021

```python
df = df[['city', 'newCases']]
df10 = df.nlargest(10, 'newCases')
df10
```

```python
df10.plot.bar('city', 'newCases', title='10 cidades com mais novos casos em 31/7/2021')
plt.xlabel('Cidades')
plt.ylabel('Número de novos casos')
```
